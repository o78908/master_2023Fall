# dfs

```python
from collections import deque

def bfs(n, m, edges, s):
    graph = [[] for _ in range(n)]
    for u, v in edges:
        graph[u].append(v)
        graph[v].append(u)
    dist = [-1] * n
    dist[s] = 0
    q = deque([s])
    while q:
        u = q.popleft()
        for v in graph[u]:
            if dist[v] == -1:
                dist[v] = dist[u] + 1
                q.append(v)
    return dist

def main():
    n, m = map(int, input().split())
    edges = [tuple(map(int, input().split())) for _ in range(m)]
    s = int(input())
    dist = bfs(n, m, edges, s)
    for i in range(n):
        if i != s:
            print(dist[i], end=' ')
    print()

if __name__ == '__main__':
    main()
```

# BFS in 2D matrix

```python
from collections import deque

def bfs(grid, i, j):
    n, m = len(grid), len(grid[0])
    q = deque([(i, j)])
    grid[i][j] = 0
    ans = 1
    while q:
        i, j = q.popleft()
        for di, dj in ((-1, 0), (1, 0), (0, -1), (0, 1)):
            ni, nj = i + di, j + dj
            if 0 <= ni < n and 0 <= nj < m and grid[ni][nj] == 1:
                grid[ni][nj] = 0
                ans += 1
                q.append((ni, nj))
    return ans

def main():
    n, m = map(int, input().split())
    grid = [list(map(int, input().split())) for _ in range(n)]
    ans = 0
    for i in range(n):
        for j in range(m):
            if grid[i][j] == 1:
                ans = max(ans, bfs(grid, i, j))
    print(ans)

if __name__ == '__main__':
    main()
```

# dfs in 2D matrix

```python
def dfs(grid, i, j):
    if i < 0 or i >= len(grid) or j < 0 or j >= len(grid[0]) or grid[i][j] == 0:
        return 0
    grid[i][j] = 0
    return 1 + dfs(grid, i - 1, j) + dfs(grid, i + 1, j) + dfs(grid, i, j - 1) + dfs(grid, i, j + 1)

def main():
    n, m = map(int, input().split())
    grid = [list(map(int, input().split())) for _ in range(n)]
    ans = 0
    for i in range(n):
        for j in range(m):
            ans = max(ans, dfs(grid, i, j))
    print(ans)

if __name__ == '__main__':
    main()
```

```python
'''389. Find the Difference
Given two strings s and t which consist of only lowercase letters.
String t is generated by random shuffling string s and then add one more letter 
at a random position. Find the letter that was added in t.
Input:
s = "abcd"
t = "abcde"
Output: e ('e' is the letter that was added.)
'''
def findTheDifference(self, s, t):
    from collections import Counter
    s_counter = Counter(s)
    t_counter = Counter(t)
    return list(t_counter - s_counter)[0]
```

```python
'''819. Most Common Word
Given a paragraph and a list of banned words, return the most frequent word 
that is not in the list of banned words. It is guaranteed there is at least 
one word that isn't banned, and that the answer is unique.
Words in the list of banned words are given in lowercase, and free of punctuation.
Words in the paragraph are not case sensitive.  The answer is in lowercase.
Example:
Input: 
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit."
banned = ["hit"]
Output: "ball"
'''
def mostCommonWord(self, paragraph, banned):
    import re
    from collections import Counter
    # 取代標點符號
    words = re.sub("[!|?|'|,|;|.]", '', paragraph).lower().split(' ')
    return (Counter(word for word in words if word not in banned).most_common(1)[0][0])
```

# deque
deque([iterable[, maxlen]])  
enqueue參數 queue = deque(['name','age','DOB']) 裡面放幾個元素都可以  
append(x):將 x 自 deque 的右側加入。  
appendleft(x):將 x 自 deque 的左側加入。  
clear():將所有元素從 deque 中移除，使其長度為 0。  
copy():建立一個 deque 的淺複製 (shallow copy)。  
count(x):計算 deque 內元素為 x 的個數。  
extend(iterable):將 iterable 引數加入 deque 的右側。  
extendleft(iterable):將 iterable 引數加入 deque 的左側。要注意的是，加入後的元素順序和 iterable 參數是相反的。  
index(x[, start[, stop]])  
回傳 deque 中 x 的位置（或在索引 start 之後、索引 stop 之前的位置）。回傳第一個匹配的位置，如果沒找到就引發 ValueError。  
insert(i, x):在 deque 位置 i 中插入 x。如果此插入操作導致 deque 超過其長度上限 maxlen 的話，會引發 IndexError 例外。  
pop():移除並回傳 deque 的最右側元素，若本來就沒有任何元素，則會引發 IndexError。  
popleft():移除並回傳 deque 的最左側元素，若本來就沒有任何元素，則會引發 IndexError。  
remove(value):移除第一個出現的 value，如果沒找到的話就引發一個 ValueError。  
reverse():將 deque 中的元素原地 (in-place) 倒序排列並回傳 None。  
rotate(n=1):將 deque 向右輪轉 n 步。若 n 為負值則向左輪轉。當 deque 不是空的，向右輪轉一步和 d.appendleft(d.pop()) 有相同意義，而向左輪轉亦等價於 d.append(d.popleft())。

```python
from collections import deque
d = deque('ghi')                 # make a new deque with three items
# deque是可迭代的
for elem in d:                   # iterate over the deque's elements
    print(elem.upper())

#append加元素
d.append('j')                    # add a new entry to the right side
d.appendleft('f')                # add a new entry to the left side
print(d)                                # show the representation of the deque
#deque(['f', 'g', 'h', 'i', 'j'])

#pop出元素
d.pop()                          # return and remove the rightmost item
#'j'
d.popleft()                      # return and remove the leftmost item
#'f'
list(d)                          # list the contents of the deque
#['g', 'h', 'i']
d[0]                             # peek at leftmost item
#'g'
d[-1]                            # peek at rightmost item
#i

# 可以reverse用list包
list(reversed(d))                # list the contents of a deque in reverse
#['i', 'h', 'g']

# 判斷在不在裡面
'h' in d                         # search the deque
#True


# extend加元素
d.extend('jkl')                  # add multiple elements at once
d
#deque(['g', 'h', 'i', 'j', 'k', 'l'])

# 往右轉，最右邊會跑到最左邊
d.rotate(1)                      # right rotation
print(d)
#deque(['l', 'g', 'h', 'i', 'j', 'k'])
d.rotate(-1)                     # left rotation
print(d)
#deque(['g', 'h', 'i', 'j', 'k', 'l'])

deque(reversed(d))               # make a new deque in reverse order
deque(['l', 'k', 'j', 'i', 'h', 'g'])
d.clear()                        # empty the deque
d.pop()                          # cannot pop from an empty deque
d.extendleft('abc')              # extendleft() reverses the input order
d
#deque(['c', 'b', 'a'])
```

# [Maximum Palindromes](https://www.hackerrank.com/challenges/maximum-palindromes/problem?isFullScreen=true  ) 

Your job as the head of the marketing team is to answer all the queries. Since the answers can be very large, you are only required to find the answer modulo 10^9+7  
Complete the functions initialize and answerQuery and return the number of maximum-length palindromes modulo 10^9+7

```python
import math
import os
import random
import re
import sys
from collections import Counter, defaultdict
from math import factorial
# O(k * log n)
#
# Complete the 'initialize' function below.
#
# The function accepts STRING s as parameter.
#
fact = dict()
powr = dict()
dist = defaultdict(lambda : Counter(""))
# m 被定義為 10 ** 9 + 7，這是一個常見的質數，通常在計算中使用，以防止數值溢出或過大。取模運算有時可以確保計算的結果保持在一個較小的範圍內，避免數值過大而導致溢出或不穩定的計算。
#% m 部分確保計算結果的範圍在 0 到 m-1 之間，以避免可能的數值溢出
m = 10 ** 9 + 7

def initialize(s):
    # This function is called once before all queries.
    # fact 是用來保存階乘的，powr 是用來保存冪次的，而 dist 是一個字典，每個鍵的值是一個 Counter，用來記錄字符串中每個位置之前的字符分佈
    # Counter(s[0])字符串中第一個位置的字符分,e.g abba  Counter({'a': 1})
    # 任何樹的階乘0都是1，次方0都是1
    fact[0], powr[0], dist[0] = 1, 1, Counter(s[0])
    for j in range(1, len(s)):
        # fact[j] = (j * fact[j - 1]) % m 的目的是計算 j 的階乘 (j * fact[j - 1])，然後將結果對 m 取餘數，以避免數字變得太大。這在取模運算時是一種常見的做法，特別是當計算的中間結果可能超過整數表示範圍時。
        fact[j] = (j * fact[j - 1]) % m
        # 更新 dist[j]，將 s[j] 的字符添加到前一個位置的字符分佈中。這樣 dist 中的每個鍵對應的值就是該位置之前的字符分布
        # dist[j] 的目的是記錄字符串中每個位置之前的字符分佈情況。
        dist[j] = dist[j-1] + Counter(s[j])

# O(log n)
def power(x, n, m):
    if n == 1:
        return x % m
    elif n % 2 == 0:
        return power(x ** 2 % m, n // 2, m)
    else:
        return (x * power(x ** 2 % m, (n - 1) // 2, m)) % m

#
# Complete the 'answerQuery' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER l
#  2. INTEGER r
#
#O(k)
def answerQuery(l, r):
    # b = dist[r-1] - dist[l-2]: 這一行計算了在位置 r-1 的字符分佈和在位置 l-2 的字符分佈之間的差異。這是透過 dist 字典的前綴和來實現的。dist[r-1] 表示位置 r-1 之前的字符分佈，而 dist[l-2] 表示位置 l-2 之前的字符分佈。
    b = dist[r-1] - dist[l-2]
    # 計算最大長度迴文子串的數量, 可以形成迴文對的字符對數量（count）和可以單獨形成迴文的字符數量（p)
    p, count, value = 0, 0, 1
     #看下面
    for c in b.values():
        if c >= 2:
            # 將出現次數除以 2的整數部分加到 count 中，表示這些字符可以形成的迴文對數
            count += c // 2
            # 更新 value，這是一個階乘的累積，表示每種字符形成迴文的方式的組合數
            value = (value * fact[c // 2]) % m
        if c % 2 == 1:
            # 表示這個奇數次數的字符可以單獨形成迴圈
            p += 1
    # max(1, p): 保證至少有一種字符可以形成迴文。
    # fact[count]: 考慮到每種字符形成迴文的方式的組合數。
    return (max(1, p) * fact[count] * power(value, m - 2, m)) % m
    # Return the answer for this query modulo 1000000007.
    
    
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    initialize(s)

    q = int(input().strip())

    for q_itr in range(q):
        first_multiple_input = input().rstrip().split()

        l = int(first_multiple_input[0])

        r = int(first_multiple_input[1])

        result = answerQuery(l, r)

        fptr.write(str(result) + '\n')

    fptr.close()
```

使用階乘的主要原因是排列組合的計算。如果我們有 c 個相同的元素，想要從中選取 c // 2 個元素，不考慮它們的排列順序，有多少種不同的方式呢？這就是排列組合中的組合數。  
階乘的公式是一種用於計算排列組合的工具。具體而言，對於正整數 n，n 的階乘（n!）表示所有小於等於 n 的正整數的乘積。在這個問題中，我們使用 fact[c // 2] 來表示選取 c // 2 個元素的組合數。  
(x^n) % m，保存冪次的列表 power 就可以用來儲存先前計算過的冪次，以減少冪次計算的時間複雜度

dist[j] = dist[j-1] + Counter(s[j])  
這行程式碼的目的是更新 dist[j]，即在字典 dist 中新增一個鍵值對，其中鍵是位置 j，值是 Counter(s[j])，表示字符串 s 中位置 j 之前的字符分佈。  

假設 s = "abca"，一開始 dist[0] 被初始化為 Counter("a")，表示位置 0 之前的字符分佈。接著，我們進行迴圈遍歷 s 的每個位置 j：  
當 j = 1 時，我們計算 Counter(s[j])，即 Counter("b")，然後將其加到 dist[j-1] 上。現在，dist[1] 表示位置 1 之前的字符分佈，即 Counter("ab")。  
當 j = 2 時，我們計算 Counter(s[j])，即 Counter("c")，然後將其加到 dist[j-1] 上。現在，dist[2] 表示位置 2 之前的字符分佈，即 Counter("abc")。  
當 j = 3 時，我們計算 Counter(s[j])，即 Counter("a")，然後將其加到 dist[j-1] 上。現在，dist[3] 表示位置 3 之前的字符分佈，即 Counter("abca")。  
最終，dist 的值會隨著迴圈的進行而更新，每個位置 j 對應的 dist[j] 都表示該位置之前的字符分佈。  

x^n 可以分解為 (x^2)^(n/2)  
偶數情況： 如果指數 n 為偶數，則遞歸計算 x 的一半的平方，這樣的遞歸結構利用了冪次方的性質，即 x^n = (x^2)^(n/2)。  
奇數情況： 如果指數 n 為奇數，則遞歸計算 x 的一半的平方，最後再乘上 x x^n = x * (x^2)^((n-1)/2)

因為 n 不等於 1，而且 n 是偶數，所以再次進入 elif 分支：  
計算 x ** 2 % m，即 4 ** 2 % 5 = 16 % 5 = 1。  
遞迴調用 power(1, 1, 5)。

計算 (x * power(x ** 2 % m, (n - 1) // 2, m)) % m：  
計算 x ** 2 % m，即 1 ** 2 % 5 = 1 % 5 = 1。  
遞迴調用 power(1, 0, 5)。  
因為 n 等於 0，所以直接返回 1 % 5 = 1。  
將計算的結果返回。

(max(1, p) * fact[count] * power(value, m - 2, m)) % m  
p: 這個變數用於追蹤在範圍 [l, r] 內字符出現奇數次的情況。在後續的程式中，如果某個字符出現奇數次，那麼這個字符可以單獨形成一個迴文，因此 p 用於計數這樣的字符的數量。  
count: 這個變數用於追蹤在範圍 [l, r] 內字符出現偶數次的情況。在後續的程式中，如果某個字符出現偶數次，那麼這個字符可以用來形成迴文對，因此 count 用於計數這樣的字符的對數。  
value: 這是一個累積的變數，用於計算字符出現偶數次時形成迴文的方式的組合數。在後續的程式中，這個變數會被更新，以考慮每種字符形成迴文的方式的組合數。

```python
from collections import deque

# 先把list轉tuple，因為list可變 tuple不可變
def tuplify(x):
    return tuple(tuple(i) for i in x)

def moves(rods):
    # 遞迴柱子
    for x in range(4):
        # 檢查柱子 x 是否有盤子，即 rods[x] 是否不為空
        if rods[x]:
            # 對於每一個柱子 x，迭代變數 y 從 0 到 3，這代表著可以將盤子移動到的四個目標柱子的編號
            for y in range(4):
                # 檢查目標柱子 y 是否為空，或者檢查目標柱子 y 最上面的盤子是否大於柱子 x 最上面的盤子。這確保只有當目標柱子是空的，或者目標柱子的最上面的盤子大於要移動的柱子 x 的最上面的盤子時，才生成合法的移動
                if not rods[y] or rods[y][-1] > rods[x][-1]:
                    # 如果上述條件滿足，則使用 yield 生成一個包含目前柱子編號 x 和目標柱子編號 y 的元組，表示可以從柱子 x 移動到柱子 y
                    yield (x, y)

def do_move(rods, x, y):
    # 複製柱子，避免直接修改柱子
    rods = [list(r) for r in rods]
    # 將柱子 x 最上面的盤子移動到柱子 y
    # 這裡使用 pop() 移除柱子 x 最上面的盤子，然後使用 append() 將該盤子加入柱子 y
    rods[y].append(rods[x].pop())
    # 根據每根柱子最上面的盤子大小進行排序
    # lambda t: t[-1] if t else 0 柱子最上面的盤子的大小，如果柱子為空，則返回 0。排序後，這確保柱子上的盤子按從大到小的順序排列
    rods[1:4] = sorted(rods[1:4], key=lambda t: t[-1] if t else 0)
    # 把用好的柱子存成tuple
    return tuplify(rods)

def bfs(rods, n):
    start = (tuplify(rods), 0)
    visited = set()
    visited.add(start)
    q = deque([start])
    while q:
        # cur, depth = q.popleft()：彈出隊列中的一個元素，其中 cur 是當前的狀態
        cur, depth = q.popleft()
        # 檢查是否已達到目標狀態，即所有柱子上都沒有盤子。如果是，則返回深度
        if all(len(r) == 0 for r in cur[1:]):
            return depth
        # 遞迴所有合法的移動
        for x, y in moves(cur):
            child = do_move(cur, x, y)
            if child not in visited:
                visited.add(child)
                q.append((child, depth + 1))
    return -1

n = int(input())
rods = [[] for _ in range(4)]
for i, disk in enumerate(map(int, input().split())):
    # 將每個盤子放到相應的柱子上。disk-1 是因為柱子編號是從 1 開始的，而列表的索引是從 0 開始的
    rods[disk-1] = [i] + rods[disk-1]
print(bfs(rods, n))
```

螃蟹的組成：

每個螃蟹的腳部節點到超級匯點的邊的容量為1，表示每條腳只能連接到一個頭部。  
在這個問題中，要求找到一些互不相交的螃蟹形狀的子圖，並使得這些子圖的總節點數最大。這裡的子圖必須滿足以下條件：  
1.每個子圖都是一個螃蟹，擁有一個頭部和固定數量的腳部，且這些腳部與頭部相連。  
2.任意兩個子圖之間沒有共同的節點，即它們是互不相交的。  

具體而言，每個測試用例的輸入包括：

t：每個螃蟹的最大腳部數量

接下來的m行，每行包含兩個空格分隔的數值 v1i 和 v2i，表示圖中存在一條邊，連接著節點 v1i 和 v2i。  

解決問題的目標是設計一個演算法，從給定的圖中選擇螃蟹形狀的子圖，使得這些子圖的總節點數最大，同時保持它們是互不相交的。程式碼中使用了最大流算法的實現，通過找到互不相交的增廣路徑，來達到最大化節點數的目標。

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'crabGraphs' function below.
#
# The function is expected to return an INTEGER.
# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER t
#  3. 2D_INTEGER_ARRAY graph
#
from math import inf
import queue
def bfs(G, parent, s, t):
    n = len(G)
    visited = [False for _ in range(n)]
    q = queue.Queue(maxsize=0)
     # 將起始節點 s 放入 queue 中，並將其標記為已訪問
    q.put(s)
    visited[s] = True
    while not q.empty():
        # 從 queue 中取出一個節點 curr
        curr = q.get()
        # 對於 curr 的每一個鄰接節點 v，如果 v 沒有被訪問過，並且 v 的容量大於 0，則將 v 放入 queue 中，並將 v 標記為已訪問，同時將 v 的父節點設為 curr
        for v, val in enumerate(G[curr]):
            if not visited[v] and val > 0:
                q.put(v)
                visited[v] = True
                parent[v] = curr
    # 如果終點節點 t 被訪問過，則返回 True，否則返回 False
    return visited[t]

# G: 圖的鄰接矩陣, S: 源點, T: 決點
def maxFlow(G, s, t):
    # n 是圖中節點的個數
    n = len(G)
    # parent 用來記錄路徑，parent[v] 表示從源點 s 到節點 v 的路徑上 v 的父節點
    parent = [None for _ in range(n)]
    max_flow = 0
    # 當還存在BFS時，不斷對圖進行 BFS
    while bfs(G, parent, s, t):
        # path_flow 用來記錄路徑上的最小容量，初始值為無限大
        path_flow = inf
        # 由於 parent 記錄的是從源點 s 到節點 t 的路徑，因此可以通過不斷訪問 parent[v] 來找到從源點 s 到節點 t 的路徑上的所有節點
        v = t
        # 通過訪問 parent[v]，找到從源點 s 到節點 t 的路徑上的最小容量
        while v != s:
            # G[parent[v]][v] 找到從源點 s 到節點 v 的路徑上 v 的容量, 並更新 path_flow
            path_flow = min(path_flow, G[parent[v]][v])
            # 將 v 設為父節點，繼續尋找從源點 s 到節點 t 的路徑上的最小容量
            v = parent[v]
        # 將從源點 s 到節點 t 的路徑上的最小容量加到 max_flow 中
        max_flow += path_flow
        # 通過訪問 parent[v]，更新從源點 s 到節點 t 的路徑上的所有節點的容量
        v = t
        # 通過訪問 parent[v]，更新從源點 s 到節點 t 的路徑上的所有節點的容量
        while v != s:
            # u 是 v 的父節點，通過更新 G[u][v] 和 G[v][u] 來更新從源點 s 到節點 t 的路徑上的所有節點的容量
            u = parent[v]
            # 正向邊的容量減去 path_flow
            # 考慮一條正向邊 (u, v)，這代表從節點 u 到節點 v 的一條通路。我們將 path_flow 的值減去這條邊，以表示我們將流量 path_flow 從 u 流向 v。這樣，正向邊上的剩餘容量將減少。
            G[u][v] -= path_flow
            # 反向邊的容量加上 path_flow
            # 考慮一條反向邊 (v, u)，這代表從節點 v 到節點 u 的一條通路。我們將 path_flow 的值加上這條邊，以表示我們將流量 path_flow 從 v 流向 u。這樣，反向邊上的剩餘容量將增加。
            G[v][u] += path_flow
            v = parent[v]
    return max_flow


def crabGraphs(n, T, graph):
    # 將圖中的節點分成兩部分，前 n 個節點是左邊的節點，後 n 個節點是右邊的節點
    #(2*n + 2) * (2*n + 2) 的矩陣，其中 2*n 是因為左邊的節點和右邊的節點都有 n 個，另外 2 是因為源點和決點
    G = [[0 for _ in range(2*n + 2)] for _ in range(2*n+2)]
    # 將左邊的節點和右邊的節點分別標號為 0 到 n-1 和 n 到 2n-1
    for u, v in graph:
        # 對於每一條邊 (u, v)，在圖中添加一條從 u 到 v 的邊和一條從 v 到 u 的邊，容量均為 1
        # 這裡使用 u-1 和 v-1 是因為節點的編號是從1開始的，而在矩陣中的索引是從0開始的
        #  s 到這些頭部節點的容量設置為不超過 T，並且不超過與其相連的腳部節點的總和
        G[u-1][n+v-1] = 1
        G[v-1][n+u-1] = 1
    # 將源點 s 標號為 2n，將決點 t 標號為 2n+1
    s = 2*n
    t = 2*n + 1
    # 將源點 s 和決點 t 與圖中的節點相連，容量均為 T
    for u in range(n):
        #  min(T, sum(G[u])) 保證了頭部最多只能連接 T 條腳，即不超過每個螃蟹頭的容量限制
        G[s][u] = min(T, sum(G[u]))
        # 這樣每個螃蟹的腳部節點只能連接一條邊到超級匯點，這保證了每條腳都能被利用
        G[n+u][t] = 1
    return maxFlow(G, s, t)

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    c = int(input().strip())

    for c_itr in range(c):
        first_multiple_input = input().rstrip().split()

        n = int(first_multiple_input[0])

        t = int(first_multiple_input[1])

        m = int(first_multiple_input[2])

        graph = []

        for _ in range(m):
            graph.append(list(map(int, input().rstrip().split())))

        result = crabGraphs(n, t, graph)

        fptr.write(str(result) + '\n')

    fptr.close()
```

# [Red Knight's Shortest Path](https://www.hackerrank.com/challenges/red-knights-shortest-path/problem?isFullScreen=true)

In ordinary chess, the pieces are only of two colors, black and white. In our version of chess, we are including new pieces with unique movements. One of the most powerful pieces in this version is the red knight.  
The red knight can move to six different positions based on its current position (UpperLeft, UpperRight, Right, LowerRight, LowerLeft, Left) as shown in the figure below.
image  

The board is a grid of size nXn . Each cell is identified with a pair of coordinates(i, j) , where  i is the row number and j is the column number, both zero-indexed. Thus,(0,0)  is the upper-left corner and  is the bottom-right corner.  

Complete the function $\color{red}{printShortestPath}$, which takes as input the grid size n , and the coordinates of the starting and ending position (istart, jstart) and  respectively(iend, jend), as input. The function does not return anything.  

Given the coordinates of the starting position of the red knight and the coordinates of the destination, print the minimum number of moves that the red knight has to make in order to reach the destination and after that, print the order of the moves that must be followed to reach the destination in the shortest way. If the destination cannot be reached, print only the word "Impossible".  

Note: There may be multiple shortest paths leading to the destination. Hence, assume that the red knight considers its possible neighbor locations in the following order of priority: UL, UR, R, LR, LL, L. In other words, if there are multiple possible options, the red knight prioritizes the first move in this list, as long as the shortest path is still achievable. Check sample input 2 for an illustration.  

### 解題邏輯
這題一樣BFS，BFS要輸出5個參數(i_start, j_start, i_end, j_end, n)給def printShortestPath(n, i_start, j_start, i_end, j_end)   
要判斷可以走哪裡，所以要另外輸出一個def check_pairs(i, j, n) 只能走UL(i-2, j-1), UR(i-2, j+1), R(i, j+2), L(i, j-2) LR(i+2, j+1) LL(i+2, j-1)

```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import deque

#
# Complete the 'printShortestPath' function below.
#
# The function accepts following parameters:
#  1. INTEGER n
#  2. INTEGER i_start
#  3. INTEGER j_start
#  4. INTEGER i_end
#  5. INTEGER j_end
#

def printShortestPath(n, i_start, j_start, i_end, j_end):
    # Print the distance along with the sequence of moves.
    def check_pairs(i, j, n):
        # 6個可能的移動
        possible_pairs = [(i-2, j-1, 'UL'), (i-2, j+1, 'UR'), (i, j+2, 'R'), (i, j-2, 'L'), (i+2, j+1, 'LR'), (i+2, j-1, 'LL')]
        # 用來存放可以走的移動
        moves = {}
        for x in possible_pairs:
            # X[0]是i, X[1]是j, X[2]是移動方向
            if 0 <= x[0] < n and 0 <= x[1] < n:
                # 如果可以走，就把移動方向加到moves
                moves[x] = 1
        return moves
        
    def bfs(a, b, x, y, n):
        # deque a是i, b是j, 0是count, []是path
        q = deque([(a, b, 0, [])])
        # 已經走過設為1
        visited = {(a, b): 1}
        while q:
            # 從左邊取出一個元素
            i, j, count, path = q.popleft()
            # 檢查是否可以走
            pos_pairs = check_pairs(i, j, n)
            # 對於每一個可以走的移動
            for pair in pos_pairs:
                # pair[0]是i, pair[1]是j, pair[2]是移動方向
                coord = (pair[0], pair[1])
                # new_path是目前的路徑加上移動方向
                new_path = path + [pair[2]]
                # 如果移動到終點，就回傳
                if coord == (x, y):
                    # count+1是因為從0開始算，但是要算移動次數
                    print(count+1)
                    # *new_path是把list轉成string
                    print(*new_path)
                    return
                # 如果可以走，且還沒走過
                if coord not in visited:
                    # 把visited設為1
                    visited[coord] = 1
                    # 把新的移動加到q
                    q.append((coord[0], coord[1], count+1, new_path))
        # 如果沒有路徑，就回傳Impossible
        print("Impossible")
    bfs(i_start, j_start, i_end, j_end, n)
            

if __name__ == '__main__':
    n = int(input().strip())

    first_multiple_input = input().rstrip().split()

    i_start = int(first_multiple_input[0])

    j_start = int(first_multiple_input[1])

    i_end = int(first_multiple_input[2])

    j_end = int(first_multiple_input[3])

    printShortestPath(n, i_start, j_start, i_end, j_end)
```

```python
#!/bin/python3
import random
import re
import sys

def prims(n, edges, start):
    
    # 是為了給下面用後來才加的
    # return node with minimum weight
    def findMinIndex(visited, weights):
        minWeight = sys.maxsize
        # -1 means no node found
        minIndex = -1
        # loop through all the nodes
        for i in range(1, len(weights)):
            # if node is not visited and weight is less than minWeight
            if not visited[i] and weights[i] < minWeight:
                # update minWeight and minIndex
                minWeight = weights[i]
                # update minIndex
                minIndex = i
        return minIndex
    
    # initialization
    # create adjecency matrix
    graph = [[-1 for _ in range(n+1)] for _ in range(n+1)]
    # assign weight to the edges
    for u, v, w in edges:
        # bidirectional
        graph[u][v] = w
        graph[v][u] = w
    # visited array
    visited = [False] * (n+1)
    # distance array
    weights = [sys.maxsize] * (n+1)
    weights[start] = 0

    #main logic
    # find the minimum weight vertex
    for i in range(1, n+1):
        u = findMinIndex(visited, weights)
        visited[u] = True
        for v in range(1, n+1):
            # if there is an edge between u and v and v is not visited
            # weight of v is greater than weight of u
            if graph[u][v] >= 0 and visited[v] == False and weights[v] > graph[u][v]:
                weights[v] = graph[u][v]
    return sum(weights[1:])

    

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    first_multiple_input = input().rstrip().split()

    n = int(first_multiple_input[0])

    m = int(first_multiple_input[1])

    edges = []

    for _ in range(m):
        edges.append(list(map(int, input().rstrip().split())))

    start = int(input().strip())

    result = prims(n, edges, start)

    fptr.write(str(result) + '\n')

    fptr.close()
```

# [Even Tree](https://www.hackerrank.com/challenges/even-tree/problem?isFullScreen=true)

Find the maximum number of edges you can remove from the tree to get a forest such that each connected component of the forest contains an even number of nodes.
As an example, the following tree with  nodes can be cut at most  time to create an even forest.  
  
![](https://s3.amazonaws.com/hr-assets/0/1533926256-3a1cc069a7-evenforestexb.png)  

Complete the evenForest function in the editor below. It should return an integer as described.  
evenForest has the following parameter(s):  
t_nodes: the number of nodes in the tree  
t_edges: the number of undirected edges in the tree  
t_from: start nodes for each edge  
t_to: end nodes for each edge, (Match by index to t_from.)


最多砍幾刀可以成為偶數獨立的子樹

```python
#!/bin/python3

import math
import os
import random
import re
import sys

# Complete the evenForest function below.
def evenForest(t_nodes, t_edges, t_from, t_to):
    # create adjecency list
    graph = [[] for _ in range(t_nodes+1)]
    # add edges to the graph
    for x, y in zip(t_from, t_to):
        # bidirectional
        graph[x].append(y)
        graph[y].append(x)
    # visited array, 0 means not visited
    count = [0] * (t_nodes+1)

    def dfs(u, graph, count):
        count[u]= 1 # mark node as visited
        remove_edge = 0    # count of edges to be removed
        # loop through all the adjacent nodes
        for v in graph[u]:
            # if node is not visited
            if count[v] == 0:
                remove_edge += dfs(v, graph, count)
                # add the count of adjacent node to the current node
                count[u] += count[v] 
                # if count of adjacent node is even(舉例，如果有4個子節點，就要刪掉3條邊)
                if count[v] % 2 == 0:
                    remove_edge += 1

        return remove_edge
    
    # start from node 1
    return  dfs(1, graph, count)


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    t_nodes, t_edges = map(int, input().rstrip().split())

    t_from = [0] * t_edges
    t_to = [0] * t_edges

    for i in range(t_edges):
        t_from[i], t_to[i] = map(int, input().rstrip().split())

    res = evenForest(t_nodes, t_edges, t_from, t_to)

    fptr.write(str(res) + '\n')

    fptr.close()
```

Given an array of integers and a target value, determine the number of pairs of array elements that have a difference equal to the target value.
Example   
 
K = 1
arr = [1, 2, 3, 4]
There are three values that differ by k = 1 : 2-1, 3-2, and 4-3. Return .   

Function Description  
Complete the pairs function below.  
pairs has the following parameter(s):  
  
int k: an integer, the target difference  
int arr[n]: an array of integers  
  
Returns  
int: the number of pairs that satisfy the criterion   

Input Format  
The first line contains two space-separated integers n and k, the size of arr and the target value.     
The second line contains  space-separated integers of the array arr.  

找相減＝k 且不重複  
STDIN       Function
-----       --------
5 2         arr[] size n = 5, k =2
1 5 3 4 2   arr = [1, 5, 3, 4, 2]
  
output:3 [5, 3], [4, 2], [3, 1]

```python
#!/bin/python3

import math
import os
import random
import re
import sys
from collections import Counter


def pairs(k, arr):
    # 用Counter把arr變成字典，key是數字，value是出現次數
    d = Counter(arr)
    pairs = 0
    
    # 看所有數字, 數字加上k或減去k, 有沒有在d裡面
    for i in arr:
        # i是數字，k是差距，如果i+k在d裡面，就加1
        if i + k in d:
            print(f'd={d}, i={i}, k={k}, i+k={i+k}')
            pairs += 1
        if i - k in d:
            print(f'd={d}, i={i}, k={k}, i-k={i-k}')
            pairs += 1
        # 這裡用del是因為如果有重複的數字，就會刪掉重複的數字，避免重複計算
        del d[i]
        
    return pairs

if __name__ == '__main__':
    print(pairs(4,[1,3,5,8,6,4,2,4,5,6]))
    # fptr = open(os.environ['OUTPUT_PATH'], 'w')
    # first_multiple_input = input().rstrip().split()
    # n = int(first_multiple_input[0])
    # k = int(first_multiple_input[1])
    # arr = list(map(int, input().rstrip().split()))
    # result = pairs(k, arr)
    # fptr.write(str(result) + '\n')
    # fptr.close()
```
